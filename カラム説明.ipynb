

##  データベース設計（チャットアプリ）

---

###  `users` テーブル（ユーザー情報）

| カラム名     | 型       | 説明                         |
|--------------|----------|------------------------------|
| id           | PK       | ユーザーID（自動採番）       |
| username     | 文字列   | 表示名など                   |
| email        | 文字列   | メールアドレス               |
| password     | 文字列   | ハッシュ化パスワード         |
| created_at   | 日時     | 登録日                       |

---

###  `chat_rooms` テーブル（チャットルーム）

| カラム名     | 型                 | 説明           |
|--------------|--------------------|----------------|
| id           | PK                 | ルームID       |
| name         | 文字列             | ルーム名       |
| created_by   | FK(users.id)       | 作成者         |
| created_at   | 日時               | 作成日時       |

---

###  `messages` テーブル（メッセージ）

| カラム名     | 型                     | 説明             |
|--------------|------------------------|------------------|
| id           | PK                     | メッセージID     |
| chat_room_id | FK(chat_rooms.id)      | ルームへの紐づけ |
| user_id      | FK(users.id)           | 投稿者           |
| content      | 文字列                 | メッセージ内容   |
| created_at   | 日時                   | 投稿日時         |
